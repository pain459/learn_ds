### Matplotlib example workflow

```python
# 0. import matoplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt # type: ignore

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))  # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save and show (you save the whole figure)
fig.savefig("sample-plot.png")
```

### Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Historgram
* Subplots

```python
import numpy as np # type: ignore
```

```python
# Create some data
x = np.linspace(0, 10, 100)  # return evenly spaced numbers between 0 to 10 of 100 in numbers
x[:10]
```

```python
# Plot the data and create a line plot.
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Use the same data to make scatter plot
fix, ax = plt.subplots()
ax.scatter(x, np.exp(x));  # We used exponential function here.
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));  # Create a sin wave.
```

```python
# Make a plot from a dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut Butter": 8,
                     "Cashew Butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Store",
       ylabel="Price($)");
```

```python
fix, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values());
# in some versions we have to make the keys and value as list
# ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)  # randn is sample of sample from normal distribution
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
# plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Plotting from Pandas DataFrames

```python
import pandas as pd
```

```python
# Make a dataframe
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range("1/1/2020", periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales['Price'] = car_sales['Price'].replace('[\\$\\,\\.]', '', regex=True).astype(int)
car_sales
```

```python
type(car_sales["Price"][0])
```

```python
car_sales['Price'] = (car_sales['Price'] / 100).astype(int)
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()  # we can add astype(int) before cumsum if the string ghost follows.
car_sales
```

```python
# Lets plot the total sales
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```
