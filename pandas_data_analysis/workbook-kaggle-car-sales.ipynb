```python
manufacturer_sales = df.groupby('Manufacturer')['Sales_in_thousands'].sum().reset_index()
manufacturer_sales_sorted = manufacturer_sales.sort_values('Sales_in_thousands', ascending=False).reset_index(drop=True)
manufacturer_sales_sorted
```

```python
model_sales = df.groupby('Model')['Sales_in_thousands'].sum().reset_index()
model_sales_sorted = model_sales.sort_values('Sales_in_thousands', ascending=False).reset_index(drop=True)
model_sales_sorted.head(10)
```

```python
average_price_per_model = df.groupby('Model')['Sales_in_thousands'].mean().reset_index()
average_price_per_model_sorted = average_price_per_model.sort_values('Sales_in_thousands', ascending=False).reset_index(drop=True)
average_price_per_model_sorted.head(5)
```
